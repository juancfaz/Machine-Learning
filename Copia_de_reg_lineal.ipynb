## 1. Un ejemplo en una sola dimensi칩n

Vamos a abrir y a visualizar unos datos que se encuentran en el archivo `carretas.txt` (abrelos con un editor de texto si quieres ver el archivo original). En este archivo se tiene las ganancias anuales (en dolares) de unos tacos de carreta (bueno, su equivalente gringo) respecto al tama침o de la ciudad donde se encuentra la carreta. Estos datos provienen de el curso de *Machine learning* de *coursera* de *Andrew Ng*.



```python
# Lee los datos en un nd array llamado datos

url = 'https://github.com/mcd-unison/aaa-curso/raw/main/ejemplos/carretas.txt'
datos = np.loadtxt(url, comments='%', delimiter=',')

# Separa los datos de entrada de los de salida.
# si decimos que x = datos[:,0], pues se toma solo una columna de datos,
# por lo que x ser칤a un ndarray de forma (shape) (96,). Al decir x = datos[:, 0:1]
# significa que vamos a tomar todas las columnas de 0 a una antes de 1, por lo
# que x tiene una forma (96, 1). Para mantener generalidad, es mejor manejar x como una matriz
# de una sola columna que como un vector de una dimensi칩n.
x, y = datos[:,0:1], datos[:,1]

# T es el n칰mero de instancias y n el de atributos
T, n = x.shape

plt.plot(x, y, 'rx')
plt.title(u'Ganancias anuales de una carreta de acuerdo al tama침o de una ciudad')
plt.xlabel(r"Poblaci$\'o$n ($\times 10^4$ habitantes)")
plt.ylabel(r'Beneficios ($\times 10^4$ dolares)')
plt.show()
```

Listo, ya temos los datos. La hip칩tesis que hacemos es que el valor de salida lo podemos estimar como

$$
\hat{y}^{(i)} = h_{\theta}(x^{(i)}) = \omega_1 x^{(i)} + b, \quad \theta = (b, \omega_1)
$$


El criterio MSE como funci칩n de p칠rdida es el de minimizar el costo definido como
$$
E_{in}(\omega, b) = \frac{1}{2} \sum_{i = 1}^M (y^{(i)} - \hat{y}^{(i)})^2.
$$

Por lo tanto, para saber si estamos minimizando o no, debemos ser capaces de medir la funci칩n de p칠rdida.

**Desarrolla la funci칩n de costo tal como se pide abajo**

```python
import numpy as np

def mse(x, y, w, b):
    """
    Calcula el costo de acuerdo al criterio de MSE (mean square error) asumiendo un conjunto de datos
    x, con una salida y, y una hip칩tesis lineal parametrizada por omega

    Par치metros
    ----------
    x: Un ndarray de dimension (M, n)
    y: Un ndarray de dimensi칩n (M, )
    w: Un ndarray de dimensi칩n (n, )
    b: Un flotante

    Devuelve
    --------
    Un flotante con el costo

    """
    M, n = x.shape
    loss = 0

    predictions = np.dot(x, w) + b

    squared_errors = (predictions - y) ** 2

    loss = round(1/2 * np.mean(squared_errors), 2)

    return loss
```

y para probar si est치 bien el programa, si calculamos $E_{in}(\omega, b)$ para $\omega_1 = 0$, $b = 1$ debe de dar (para este conjunto de datos) **26.73** (recuerda verificar que regrese un flotante y no un vector de un solo elemento).

```python
w = np.zeros([n])
b = 1
print("El error en muestra es {}".format(mse(x, y, w, b)))
```

Muy bien, ya podemos calcular el criterio a optimizar. Vamos entonces a utilizar la funci칩n que acabamos de hacer para ver sus valores para diferentes valores de $\omega$ y $b$.

```python
# Definimos una funci칩n de burrito (wrap) que depende solo de b y theta1
def costo_w(b, w1):
    return mse(x, y, np.array([w1]), b)

# Y ahora la convertimos en una funci칩n tipo numpy (aplica para cualquier entrada de ndarrays)
costo_vect = np.vectorize(costo_w)

#Ahora generamos la lista de valores para graficar
b = np.linspace(-15, 10, 100);
w1 = np.linspace(-2, 4, 100);

# Y los convertimos en matrices utilizando la funci칩n meshgrid
b, w1 = np.meshgrid(b, w1)

# Y calculamos los costos para cada par de theta0 y theta 1 con nuestra nueva funcion de costos vectorizada
J = costo_vect(b, w1)

# Y graficamos el contorno
plt.contour(b, w1, J, 80, linewidths=0.5, colors='k')
plt.contourf(b, w1, J, 80, cmap=plt.cm.rainbow, vmax=J.max(), vmin=J.min())
plt.plot([-3.63029143940436], [1.166362350335582], 'ro')
plt.colorbar()
plt.xlabel(r"$b$")
plt.ylabel(r"$\omega_1$")
plt.title(r"Funcion de perdida")
```

Ahora si, ya tenemos todo para hacer nuestra funci칩n para encontrar los par치metros que optimicen la funci칩n de costo (que como se puede ver en la superficie deber칤a de estar por donde $b$ vale entre 0 y -5 y $\omega_1$ entre 1 y 2).

**Desarrolla la funci칩n con descenso de gradiente.**

```python
def grad_costo(x, y, w, b):
    """
    Calcula el gradiente respecto a w y b de los datos existentes

    Par치metros
    ----------
    x: ndarray de dimension [M, n] con la matriz de dise침o
    y: ndarray de dimension [M,] con los valores de salida
    w: ndarray de dimension [n, ] con los pesos
    b: flotante con el sesgo

    Devuelve
    --------
    dw, db: donde dw es un vector de dimension de w con el gradiente
            de la funci칩n de costo respecto a w, y db es la derivada de la
            funcion de costo respecto a b

    """
    error = y - (x @ w + b)

    # Gradiente con respecto a w
    dw = -x.T @ error / len(y)

    # Gradiente con respecto a b
    db = -np.sum(error) / len(y)

    return dw, db
```

```python
def descenso_gradiente_lotes(x, y, w_0, b_0, nu, max_epoch):
    """
    Descenso de gradiente durante num_iter iteraciones para regresi칩n lineal

    Par치metros
    -----------
    x: ndarray de dimension [M, n] con los datos de entrada
    y: ndarray de dimension [M,] con los datos de salida
    w_0: ndarray de dimension [n, ] con los pesos iniciales
    b_0: flotante con el sesgo inicial
    nu: flotante con tama침o de paso o tasa de aprendizaje.
    max_epoch: numero de iteraciones (entero)

    Devuelve
    --------
    w, b, mse_iter: donde w y b tiene las dimensiones de w_0 y b_0 con los par치metros
                    aprendidos, mientras que mse_hist es un ndarray de dimensi칩n
                    [num_iter, 1] con el costo en cada iteraci칩n.

    """
    w, b = w_0.copy(), b_0
    mse_iter = np.zeros(max_epoch)
    M, n = x.shape

    for iter in range(max_epoch):
        dw, db = grad_costo(x, y, w, b)

        # Actualiza los pesos y el sesgo utilizando el descenso de gradiente
        w -= nu * dw
        b -= nu * db

        mse_iter[iter] = mse(x, y, w, b)

    return w, b, mse_iter
```

Y para saber si el algoritmo se programo bien, se puede probar en el problema del *food truck* y revisar si el valor de la funci칩n de p칠rdida se va reduciendo hasta estabilizarse en un m칤nimo.

```python
w_0 = np.zeros((n,))
b_0 = 0.0

iteraciones = 1500
alpha = 0.01

w, b, mse_historial = descenso_gradiente_lotes(x, y, w_0, b_0, alpha, iteraciones)


plt.plot(mse_historial, 'b')
plt.title(u'MSE por iteraci칩n')
plt.xlabel(u'iteraci칩n')
plt.ylabel(r'$Loss(\omega, b)$')
plt.figtext(x=.6, y=.6, s="Al final de las iteraciones:\n\n w1 = {}\n b    = {}".format(w[0], b))
plt.show()
```

Ahora vamos a revisar virtualmente si la estimaci칩n es una linea recta que pasa entre todos los puntos.

**Completa los pasos para realizar la estimaci칩n.**

```python
x_estimar = np.array([[4],[24]])

# Estimar los valores de salida utilizando el modelo entrenado
y_estimado = x_estimar @ w + b

print("Los valores estimados son: {}".format(y_estimado))
```

Si los valores que obtuviste son cercanos a 1 (10000 dolares) y 24.3 (243000 dolares) entonces estamos en los valores esperados. Ahora vamos a usar estos valores para graficar los datos reales y la estimaci칩n realizada:

```python
plt.plot(x[:], y, 'xr')
plt.plot(x_estimar[:,0], y_estimado, '-b')
plt.title(u'Ganancias anuales de una carreta de acuerso al tama침o de una ciudad')
plt.xlabel(r"Poblaci$\'o$n ($\times 10^4$ habitantes)")
plt.ylabel(r'Beneficios ($\times 10^4$ dolares)')
plt.show()
```

# 2. Un ejemplo en multiples dimensiones

Como el algortimo realizado ya funciona para muchas dimensiones, no se espera tener mucho problema para utilizarlos. As칤 que ahora vamos a cargar datos y vamos a graficar la salida respecto a dos variables

```python
url = "https://github.com/mcd-unison/aaa-curso/raw/main/ejemplos/casas_portland.txt"
datos = np.loadtxt(url, comments='%', delimiter=',')
x, y = datos[:, :-1], datos[:,-1]

# M es el n칰mero de instancias y n el de atributos
M, n = x.shape

plt.plot(x[:,0], y, 'rx')
plt.title(u'Costo de una casa en relaci칩n a su tama침o')
plt.xlabel(u"tama침o (pies cuadrados)")
plt.ylabel('costo ')
plt.figure()
plt.plot(x[:,1], y, 'rx')
plt.title(u'Costo de una casa en relaci칩n al n칰mero de cuartos')
plt.xlabel("cuartos")
```

Antes de realizar el aprendizaje podemos ver que mientras una de las variables se mide en miles de pies cuadrados, la otra variable tiene valores de 1 a 4. Esto es un problema para el algoritmo del descenso de gradiente, por lo que es necesario normalizar los datos (solo para este algoritmo) y que funcione de manera correcta.

Para normalizar requerimos de dos pasos, por un lado, obtener los valores de medias y desviaciones estandares por atributo, y en segundo lugar, realizar la normalizaci칩n. Los valores de medias y desviaciones estandares hay que guardarlos, ya que ser치n necesarios para poder normalizar los datos que se quiera estimar.

**Estas funciones ya las hicimos realizadas pero revisalas por favor.**

```python
def obtiene_medias_desviaciones(x):
    """
    Obtiene las medias y las desviaciones estandar atributo a atributo.

    Par치metros
    ----------
    x: un ndarray de dimensi칩n (T, n) donde T es el n칰mro de elementos y
       n el n칰mero de atributos

    Devuelve
    ---------
    medias, desviaciones: donde ambos son ndarrays de dimensiones (n,) con
                          las medias y las desviaciones estandar respectivamente.

    """
    medias = x.mean(axis = 0)
    desviaciones = x.std(axis = 0)
    return medias, desviaciones


def normaliza(x, medias, desviaciones):
    """
    Normaliza los datos x

    Par치metros
    ----------
    x: un ndarray de dimensi칩n (T, n) donde T es el n칰mro de elementos y n el n칰mero de atributos
    medias: ndarray de dimensiones (n,) con las medias con las que se normalizar치
    desviaciones: ndarray de dimensiones (n,) con las desviaciones con las que se normalizar치

    Devuelve
    --------
    x_norm, un ndarray de las mismas dimensiones de x pero normalizado

    """
    return (x - medias) / desviaciones


# Y ahora vamos a hacer algo muy simple para probar, que pueden corroborar con el uso de una calculadora com칰n.
x_prueba = np.array([[1, 300],
                    [3, 100],
                    [2, 400],
                    [4, 200]])

m, d = obtiene_medias_desviaciones(x_prueba)

print("Los datos son: \n{}".format(x_prueba))
print("Las medias son: \n{}".format(m))
print("Las desviaciones son: \n{}".format(d))
print("Los datos normalizados son: \n{}".format(normaliza(x_prueba, m, d)))
```

Listo, entonces ya podemos hacer descenso de gradiente, o casi. El problema es que no sabemos cual ser칤a el mejor valor para $\alpha$. Escoge el valor de $\alpha$ realizando una gr치fica de 50 iteraciones solamente para valores desde 0.001, 0.005, 0.01, 0.05, 0.1, 0.5, ... y decide cual de los valores es el que m치s te conviene.

**Selecciona un valor, especifica aqu칤 cual es, y justifica porque lo seleccionaste.**

Como podemos ver en las graficas, conforme nosotros aumenbtemos las iteraciones o epochs, cambia el error para cada alpha. Desde la iteracion n칰mero 1 el alpha = 0.5 se posiciona como la mejor opci칩n y si continuamos el error se va acercando poco a poco al 0 mientras que las dem치s tardan m치s en bajar. Por lo tanto la mejor opci칩n de alpha es 0.5

```python
medias, desviaciones = obtiene_medias_desviaciones(x)
x_norm = normaliza(x, medias, desviaciones)

w_ini = np.zeros((n,))
b_ini = 0
num_iters = [5, 10, 25, 50]

#lpha = 0.0001  # Aqui es donde hay que hacer las pruebas
alphas = [0.001, 0.005, 0.01, 0.05, 0.1, 0.5, 1.3]
mse_results = []

for alpha in alphas:
    mse_alpha = []
    for num_iter in num_iters:
        _, _, mse_hist = descenso_gradiente_lotes(x_norm, y, w_ini, b_ini, alpha, num_iter)
        mse_alpha.append(mse_hist)
    mse_results.append(mse_alpha)

# Grafica los resultados para cada combinaci칩n de alpha y num_iters
for num_iter in num_iters:
    plt.figure(figsize=(12, 6))
    for alpha in alphas:
        _, _, mse_hist = descenso_gradiente_lotes(x_norm, y, w_ini, b_ini, alpha, num_iter)
        plt.plot(mse_hist, label=f"alpha = {alpha}")

    plt.title(f"Curvas de aprendizaje para num_iters = {num_iter}")
    plt.xlabel('Iteraciones')
    plt.ylabel('MSE')
    plt.legend(loc='upper right')
    plt.show()
```

**Utilizando las iteraciones necesarias, encuentra el valor de $\omega$ y $b$ utilizando el descenso de gradiente.**

```python
num_iters = 10

alpha = 0.5

w, b, mse_hist = descenso_gradiente_lotes(x_norm, y, w_ini, b_ini, alpha, num_iters)

plt.title(r"La Curva de Aprendizaje para $\alpha =$ " + str(alpha))
plt.plot(mse_hist, '-b')
plt.xlabel('Iteraciones')
plt.ylabel('MSE')
plt.figtext(x = 0.6, y = 0.6, s = "w ={}\n b ={}".format(w[0], b))
```
