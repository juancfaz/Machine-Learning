Importamos las librerias necesarias

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import factorial
```

Datos de ejemplo

```python
X = np.array([[1, 2], [2, 3], [3, 4], [4, 5]])
y = np.array([0, 1, 0, 1])
```

Inicializamos w, b y los hiperparámetros

```python
w = np.zeros(X.shape[1])  # Inicializa w con ceros
b = 0  # Inicializa b con 0

alpha = 0.00005  # Tasa de aprendizaje
num_iterations = 2000  # Número de iteraciones del descenso de gradiente
```

Definimos el algoritmo de descenso de gradiente

```python
def gradient_descent(X, y, w, b, alpha, num_iterations):
  costs = []  # Lista para almacenar los valores de costo en cada iteración
  for i in range(num_iterations):
    theta = np.dot(X, w) + b

    dw = -np.dot(X.T, y - np.exp(theta))
    db = -np.sum(y - np.exp(theta))

    w -= alpha * dw
    b -= alpha * db
    cost = -np.sum(y * theta - np.exp(theta) - np.log(factorial(y)))
    costs.append(cost)
    print(f"Iteration {i+1}: Cost = {cost}")
  return w, b, costs
```

Entrenamos el modelo

```python
w, b, costs = gradient_descent(X, y, w, b, alpha, num_iterations)
```

Gráfica de costo vs. época

```python
print(f"Parámetros finales: w = {w}, b = {b}")

# Gráfica de costo vs. época
plt.plot(range(num_iterations), costs)
plt.xlabel('Época')
plt.ylabel('Costo')
plt.title('Costo vs. Época')
plt.show()
```
